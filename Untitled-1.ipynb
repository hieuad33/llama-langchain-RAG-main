```python
import pdfkit
import time

# Đường dẫn tới wkhtmltopdf.exe
config_path = "C:\\Program Files\\wkhtmltopdf\\bin\\wkhtmltopdf.exe"
config = pdfkit.configuration(wkhtmltopdf=config_path)

# URL của trang web
url = 'https://www.thegioididong.com/dtdd/oppo-find-n5'
# Tên file PDF output
output_pdf = 'out_with_delay.pdf'

# Các tùy chọn cho wkhtmltopdf
options = {
    'enable-local-file-access': None, # Cần thiết nếu trang web tải tài nguyên cục bộ (ít gặp)
    'enable-javascript': True,        # Đảm bảo JavaScript được bật (thường là mặc định)
    'javascript-delay': 5000,         # Đợi 5000 milliseconds (5 giây) để JS thực thi
                                      # Bạn có thể cần điều chỉnh giá trị này (tăng lên nếu chưa đủ)
    'no-stop-slow-scripts': None,     # Không dừng các script chạy chậm (có thể hữu ích)
    'load-error-handling': 'ignore',  # Bỏ qua lỗi tải tài nguyên (ví dụ: ảnh bị lỗi)
    'load-media-error-handling': 'ignore', # Bỏ qua lỗi tải media
    # 'debug-javascript': True,       # Bật debug JS (in ra console khi chạy wkhtmltopdf từ terminal)
    # 'window-status': 'ready',       # Đợi cho đến khi window.status được JS đặt thành 'ready'
                                      # (Yêu cầu JS trên trang web phải hỗ trợ)
}

print(f"Đang chuyển đổi URL: {url}")
print(f"Đợi {options.get('javascript-delay', 0)}ms cho JavaScript...")

try:
    start_time = time.time()
    pdfkit.from_url(url, output_pdf, configuration=config, options=options)
    end_time = time.time()
    print(f"Đã tạo file PDF thành công: {output_pdf}")
    print(f"Thời gian thực hiện: {end_time - start_time:.2f} giây")

except Exception as e:
    print(f"Có lỗi xảy ra: {e}")
    print("Hãy kiểm tra:")
    print(f"- Đường dẫn wkhtmltopdf có chính xác không: {config_path}")
    print(f"- Trang web {url} có truy cập được không?")
    print("- Thử tăng giá trị 'javascript-delay'.")
```

```python
import asyncio
from pyppeteer import launch
import time

async def capture_pdf_with_pyppeteer(url, output_pdf, wait_time=10, scroll_pages=3):
    """
    Chụp PDF của một trang web sử dụng Pyppeteer, đợi JS và cuộn trang.

    Args:
        url (str): URL của trang web.
        output_pdf (str): Đường dẫn file PDF output.
        wait_time (int): Thời gian (giây) đợi sau khi tải trang và sau mỗi lần cuộn.
        scroll_pages (int): Số lần cuộn trang xuống dưới để tải thêm nội dung.
    """
    browser = None
    try:
        print("Đang khởi chạy trình duyệt (Pyppeteer)...")
        browser = await launch(headless=True, args=['--no-sandbox', '--disable-setuid-sandbox'])
        page = await browser.newPage()

        print(f"Đang truy cập URL: {url}")
        # Tăng thời gian chờ timeout nếu trang tải chậm
        await page.goto(url, {'waitUntil': 'networkidle0', 'timeout': 60000}) # Đợi mạng yên tĩnh

        print(f"Đợi {wait_time} giây cho các script ban đầu...")
        await asyncio.sleep(wait_time)

        # Cuộn trang để kích hoạt tải nội dung động (lazy loading)
        if scroll_pages > 0:
            print(f"Thực hiện cuộn trang {scroll_pages} lần...")
            for i in range(scroll_pages):
                await page.evaluate('window.scrollBy(0, window.innerHeight)')
                print(f"  Đã cuộn lần {i+1}/{scroll_pages}, đợi {wait_time/2} giây...")
                await asyncio.sleep(wait_time / 2) # Đợi một chút sau mỗi lần cuộn

        # Đợi thêm lần cuối
        print(f"Đợi thêm {wait_time} giây cuối cùng trước khi in...")
        await asyncio.sleep(wait_time)

        print(f"Đang tạo file PDF: {output_pdf}")
        start_time = time.time()
        # Các tùy chọn PDF có thể tham khảo tại tài liệu của Pyppeteer/Chrome DevTools Protocol
        await page.pdf({
            'path': output_pdf,
            'format': 'A4',
            'printBackground': True, # In cả background
            # 'margin': {'top': '20px', 'right': '20px', 'bottom': '20px', 'left': '20px'},
        })
        end_time = time.time()
        print(f"Đã tạo file PDF thành công: {output_pdf}")
        print(f"Thời gian tạo PDF: {end_time - start_time:.2f} giây")

    except Exception as e:
        print(f"Có lỗi xảy ra với Pyppeteer: {e}")
    finally:
        if browser:
            print("Đang đóng trình duyệt...")
            await browser.close()

# --- Chạy hàm async ---
url_target = 'https://www.thegioididong.com/dtdd/oppo-find-n5'
pdf_output_pyppeteer = 'out_pyppeteer.pdf'

# Lưu ý: Bạn cần chạy đoạn code này trong môi trường hỗ trợ asyncio
# Ví dụ: lưu thành file .py và chạy từ terminal
if __name__ == "__main__":
   asyncio.get_event_loop().run_until_complete(
       capture_pdf_with_pyppeteer(url_target, pdf_output_pyppeteer, wait_time=5, scroll_pages=3)
   )

# Hoặc nếu dùng trong Jupyter Notebook/IPython:
# await capture_pdf_with_pyppeteer(url_target, pdf_output_pyppeteer, wait_time=5, scroll_pages=3)
```

```python
import pdfkit
import requests

def convert_html_to_pdf(html_content, output_path):
    """
    Chuyển đổi nội dung HTML thành file PDF.

    Args:
        html_content (str): Nội dung HTML cần chuyển đổi.
        output_path (str): Đường dẫn và tên file PDF đầu ra.
    """
    try:
        pdfkit.from_string(html_content, output_path)
        print(f"Đã chuyển đổi thành công nội dung HTML sang {output_path}")
    except Exception as e:
        print(f"Lỗi khi chuyển đổi sang PDF: {e}")

def get_webpage_html(url):
    """
    Lấy nội dung HTML của một trang web.

    Args:
        url (str): URL của trang web.

    Returns:
        str: Nội dung HTML của trang web, hoặc None nếu có lỗi.
    """
    try:
        # Thêm header của trình duyệt để tránh bị chặn
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        # Lấy nội dung của trang web
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Kiểm tra xem có lỗi HTTP nào không
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi lấy nội dung từ URL: {e}")
        return None
```

```python
import pdfkit
import requests
from bs4 import BeautifulSoup

def convert_webpage_to_pdf_without_css(url, output_path):
    """
    Chuyển đổi một trang web thành file PDF, bỏ qua CSS.

    Args:
        url (str): URL của trang web cần chuyển đổi.
        output_path (str): Đường dẫn và tên file PDF đầu ra.
    """
    try:
        # Thêm header của trình duyệt để tránh bị chặn
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        # Lấy nội dung của trang web
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Kiểm tra xem có lỗi HTTP nào không
        html_content = response.text

        # Sử dụng BeautifulSoup để loại bỏ các thẻ <style> và thuộc tính style inline
        soup = BeautifulSoup(html_content, 'html.parser')

        # Loại bỏ thẻ <style>
        for style in soup.find_all('style'):
            style.decompose()

        # Loại bỏ thuộc tính style inline
        for tag in soup.find_all():
            if 'style' in tag.attrs:
                del tag.attrs['style']

        # Lấy nội dung HTML đã loại bỏ CSS
        cleaned_html = str(soup)

        # Cấu hình pdfkit để không tải CSS (nếu cần thiết, mặc dù đã loại bỏ ở trên)
        path_wkhtmltopdf = r'C:\Program Files\wkhtmltopdf\bin\wkhtmltopdf.exe'  # Update the path accordingly

# Configure pdfkit to use the path to wkhtmltopdf
        config = pdfkit.configuration(wkhtmltopdf=path_wkhtmltopdf)
        options = {
            'enable-local-file-access': True, # Cho phép truy cập các file cục bộ nếu có
            'load-media-error-handling': 'skip' # Bỏ qua lỗi khi tải media (có thể liên quan đến CSS)
        }

        # Sử dụng pdfkit để chuyển đổi trực tiếp từ chuỗi HTML
        pdfkit.from_string(cleaned_html, output_path, configuration=config, options=options)
        print(f"Đã chuyển đổi thành công từ {url} sang {output_path} (bỏ qua CSS)")

    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi lấy nội dung từ URL: {e}")
    except Exception as e:
        print(f"Lỗi khi chuyển đổi sang PDF: {e}")

if __name__ == "__main__":
    # Thay URL này bằng URL của trang web bạn muốn chuyển đổi
    url_to_convert = "https://www.thegioididong.com/dong-ho-thong-minh/apple-watch-s9-41mm-vien-nhom-day-vai?utm_flashsale=1"

    # Chọn đường dẫn và tên file cho file PDF đầu ra
    output_pdf_path = "thegioididong_no_css.pdf"

    convert_webpage_to_pdf_without_css(url_to_convert, output_pdf_path)
```

```python
import requests
from bs4 import BeautifulSoup
from reportlab.pdfgen import canvas

# URL của trang web cần thu thập dữ liệu
url = "https://www.thegioididong.com/dong-ho-thong-minh/apple-watch-s9-41mm-vien-nhom-day-vai?utm_flashsale=1"

# Gửi yêu cầu GET đến trang web
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Lấy nội dung của trang web
response = requests.get(url, headers=headers)
response.raise_for_status()  # Kiểm tra xem có lỗi HTTP nào không

# Kiểm tra xem yêu cầu có thành công không
if response.status_code == 200:
    # Sử dụng BeautifulSoup để phân tích cú pháp HTML
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Trích xuất thông tin từ class "detail"
    detail_element = soup.find(class_='detail')
    detail_text = detail_element.get_text()

    # Tạo file PDF và ghi thông tin trích xuất vào đó
    pdf_filename = "product_detail_reportlab.pdf"
    c = canvas.Canvas(pdf_filename)
    c.drawString(100, 800, detail_text)
    c.save()

    print("Dữ liệu đã được lưu vào file PDF.")
else:
    print("Không thể truy cập trang web.")
```
